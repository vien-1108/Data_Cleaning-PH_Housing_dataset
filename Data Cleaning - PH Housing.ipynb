```python
# --- Extract Storeys ---
def extract_storeys(desc):
    match = re.search(r"(\d+)-storey", desc, re.IGNORECASE)
    if match:
        return int(match.group(1))
    elif "single detached" in desc.lower():
        return 1  # assume single detached = 1 storey if not specified
    return None

df["Storeys"] = df["Description"].apply(extract_storeys)

# Remove decimal values (convert to integer)
df["Storeys"] = df["Storeys"].fillna(0).astype(float).astype(int)

# Show first 10 rows
df.head(10)
```

```python
# --- Extract Project ---
import re

def extract_project(desc):
    project = None
    text = desc

    #  Priority 1: Keyword search (now includes the word before the keyword)
    match = re.search(
        r"((?:\b\w+\s+)?(?:Village|Subdivision|Homes|Residences|Estates|Camella|New City|Valley|by)[^|]*)",
        text,
        flags=re.IGNORECASE
    )
    if match:
        project = match.group(1)

    #  Priority 2: "at"
    elif re.search(r"\b at \b", text, flags=re.IGNORECASE):
        project = re.split(r"\b at \b", text, 1, flags=re.IGNORECASE)[-1]

    #  Priority 3: "|"
    elif "|" in text:
        project = text.split("|", 1)[-1]

    if project:
        # Clean trailing model names or unnecessary text
        project = re.sub(r"\|.*", "", project)
        project = re.sub(r"(?i)\bmodel\b.*", "", project)
        project = project.strip()
        return project

    return None

df["Project"] = df["Description"].apply(extract_project)

# Clean leading "in", "at", or "by" from the extracted project column
df["Project"] = df["Project"].str.replace(r"^(?:in|at|by)\s+", "", regex=True, case=False)

# Show first 20 rows
df.head(20)
```

```python
import pandas as pd
import re

# Load dataset
df = pd.read_csv("PH_housing_cleaned4.csv")

# --- Extract Property Type ---
def extract_property_type(desc):
    if "townhouse" in desc.lower():
        return "Townhouse"
    elif "single detached" in desc.lower():
        return "Single Detached"
    elif "house and lot" in desc.lower():
        return "House and Lot"
    elif "house & lot" in desc.lower():
        return "House and Lot"
    elif "residential" in desc.lower():
        return "Residential House"
    else:
        return "Other"

df["Property Type"] = df["Description"].apply(extract_property_type)


# --- Extract Storeys ---
def extract_storeys(desc):
    match = re.search(r"(\d+)-storey", desc, re.IGNORECASE)
    if match:
        return int(match.group(1))
    elif "single detached" in desc.lower():
        return 1  # assume single detached = 1 storey if not specified
    return None

df["Storeys"] = df["Description"].apply(extract_storeys)

# Remove decimal values (convert to integer)
df["Storeys"] = df["Storeys"].fillna(0).astype(float).astype(int)


# --- Extract Project ---
def extract_project(desc):
    project = None
    text = desc

    #  Priority 1: Keyword search (now includes the word before the keyword)
    match = re.search(
        r"((?:\b\w+\s+)?(?:Village|Subdivision|Homes|Residences|Estates|Camella|New City|Valley|by)[^|]*)",
        text,
        flags=re.IGNORECASE
    )
    if match:
        project = match.group(1)

    #  Priority 2: "at"
    elif re.search(r"\b at \b", text, flags=re.IGNORECASE):
        project = re.split(r"\b at \b", text, 1, flags=re.IGNORECASE)[-1]

    #  Priority 3: "|"
    elif "|" in text:
        project = text.split("|", 1)[-1]

    if project:
        # Clean trailing model names or unnecessary text
        project = re.sub(r"\|.*", "", project)
        project = re.sub(r"(?i)\bmodel\b.*", "", project)
        project = project.strip()
        return project

    return None

df["Project"] = df["Description"].apply(extract_project)

# Clean leading "in", "at", or "by" from the extracted project column
df["Project"] = df["Project"].str.replace(r"^(?:in|at|by)\s+", "", regex=True, case=False)



# Save the cleaned dataset
df.to_csv("PH_housing_cleaned5.csv", index=False)
print("Columns 'Property Type', 'Storeys', and 'Project' successfully added.")

# Show first 10 rows
df.head(10)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("PH_housing_cleaned6.csv")

# Ensure Price is numeric
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# Format for display
df["Price"] = df["Price"].apply(lambda x: f"{x:,.2f}" if pd.notna(x) else "")

# Rename 'Price' column to 'Price (PHP)'
df = df.rename(columns={"Price": "Price (PHP)"})

# Save the cleaned dataset
df.to_csv("PH_housing_cleaned_FINAL.csv", index=False)

print(df.head())

# Final Remarks
print(" ")
print(" ")

print(" Data cleaning complete!")
print(" All columns are cleaned, renamed, and arranged.")
print(" Price column formatted as 'Price (PHP)'.")
print(" Dataset is now ready for analysis and visualization! ")
```
